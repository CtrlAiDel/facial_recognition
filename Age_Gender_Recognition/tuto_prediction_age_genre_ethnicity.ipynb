## Pr√©dicition du genre, de l'√¢ge et de l'ethnicit√©

Ouais je sais tu t'en balec de l'ethnicit√© mais c'est aussi pour te montrer comment on fait pour plusieurs classes, et √ßa te permettra d'adapter aux deux pr√©dicitions que tu veux √ßa peut faire un bon exo (genre je suis ton prof tsais)

```python
X = np.array(df['pixels'].tolist())

y = df[['age', 'gender', 'ethnicity']].values
```

Tout pareil

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

Alors l√† √ßa va √™tre un peu diff√©rent, la premi√®re partie du mod√®le est exactement la m√™me, tout ce qui change est la sortie, ici on aura trois couches de sorties concat√©n√©es pour 3 pr√©dicitions diff√©rentes, les deux premi√®res c'est exactement ce qu'on a vu avant, et la derni√®re y a 5 neurones pour 5 classes et une fonction softmax (comme sigmoid mais normalis√©e)

Autre diff√©rence, la syntaxe est un peu diff√©rente mais √ßa fonctionne de la m√™me mani√®re (pour te donner les deux types d'√©criture en tensorflow)

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import callbacks
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Input


input_shape = (200,200,3)

input_model = Input(input_shape)

x = Conv2D(filters=32, kernel_size=(5, 5), padding='valid', input_shape=input_shape, activation='relu')(input_model)
x = MaxPooling2D(pool_size=(2, 2))(x)
# x = Dropout(rate=0.2)(x)

x = Conv2D(filters=64, kernel_size=(3, 3), padding='valid', activation='relu')(x)
x = MaxPooling2D(pool_size=(2, 2))(x)

x = Conv2D(filters=128, kernel_size=(3, 3), padding='valid', activation='relu')(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
# x = Dropout(rate=0.2)(x)

x = Flatten()(x)

x = Dense(units=128, activation='relu')(x)
x = Dropout(rate=0.2)(x)
x = Dense(units=64, activation='relu')(x)

age = Dense(units=1, activation='sigmoid')(x)
gender = Dense(units=1, activation='sigmoid')(x)
ethnicity = Dense(units=5, activation='softmax')(x)

model = Model(inputs=input_model, outputs=tf.concat([age, gender, ethnicity], axis=-1))


model.summary()
```

Avant de compiler, on d√©finit une nouvelle fonction de perte g√©n√©rale qui correspond √† la somme des fonctions de pertes des trois pr√©dictions pond√©r√©es en fonction de ce qui nous semble le plus important. On pourrait mettre plus de poids sur l'age comme c'est le plus galere mais ca marche pas de fou la donc j'ai un peu fait empiriquement.

```python
import tensorflow as tf


lambda_age = 1
lambda_gender = 1
lambda_ethnicity = 1

def loss_age(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true-y_pred), axis=-1)

def loss_gender(y_true, y_pred):
    return tf.keras.losses.BinaryCrossentropy()(y_true, y_pred)

def loss_ethnicity(y_true, y_pred):
    return tf.keras.losses.SparseCategoricalCrossentropy()(y_true, y_pred)


def loss_function(y_true, y_pred):
    # return loss_age(y_true[:,0], y_pred[:,0]), loss_gender(y_true[:,1], y_pred[:,1]), loss_ethnicity(y_true[:,2], y_pred[:,2:]) 
    return lambda_age*loss_age(tf.cast(y_true[:,0], tf.float32), y_pred[:,0]) +  lambda_gender*loss_gender(y_true[:,1], y_pred[:,1]) + lambda_ethnicity*loss_ethnicity(y_true[:,2], y_pred[:,2:]) 
```

On d√©finit √©galement des m√©triques pour les trois diff√©rentes pr√©dictions (diff√©rence pour la nouvelle, on utilise une categorical crossentropy car on a plusieurs classes)

```python
def metric_age(y_true, y_pred):
    return tf.keras.metrics.MAE(y_true[:,0], y_pred[:,0]) * max_age

def metric_gender(y_true,y_pred):
    return tf.keras.metrics.binary_accuracy(y_true[:,1], y_pred[:,1])

def metric_ethnicity(y_true,y_pred):
    return tf.keras.metrics.sparse_categorical_accuracy(y_true[:,2], y_pred[:,2:])
```

On le compile, en choisissant bien la bonne fonction de perte (binary cross entropy pour une classif binaire) et les bonnes m√©triques

```python
model.compile(loss=loss_function, optimizer='adam', metrics = [metric_age , metric_gender, metric_ethnicity]
            )
```

Et on entra√Æne

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                              epochs=10,  batch_size=32)
```

Bon manifestement l√† √ßa s'entraine pas, je pense qu'il faut plus de donn√©es. Je regarderai √ßa demain

```python
y_pred = model.predict(X_test)
```
