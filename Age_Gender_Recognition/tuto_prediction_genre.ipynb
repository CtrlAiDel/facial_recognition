## Pr√©dicition du genre uniquement

Pareil mais la cible est le genre 

```python
X = np.array(df['pixels'].tolist())

y = df['gender'].values
```

Tout pareil

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

Tout pareil, juste l√† on a encore un neurone de sortie car c'est une classification binaire (0 ou 1) et une fonction d'activation de sortie sigmoid qui nous donne justement une probabilit√©, qui nous permet de faire la classification.

Si on avait plusieurs classes, on aurait le m√™me nombre de neurones que le nombre de classes, et on utiliserait une fonction de sortie softmax

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import callbacks
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D


input_shape = (200,200,3)

model = Sequential()


model.add(Conv2D(filters=16, kernel_size=(5, 5), padding='valid', input_shape=input_shape, activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.2))

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3, 3), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.2))

model.add(Flatten())

model.add(Dense(units=64, activation='relu'))
model.add(Dropout(rate=0.2))
model.add(Dense(units=1, activation='sigmoid'))

model.summary()
```

On le compile, en choisissant bien la bonne fonction de perte (binary cross entropy pour une classif binaire) et la m√©trique

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Et on entra√Æne

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                              epochs=10,  batch_size=32)
```

Ca s'entraine carr√©ment bien

```python
y_pred = model.predict(X_test).round()
```

```python
fig, axs = plt.subplots(4, 5, figsize=(11, 7))

for row in range(4):
    for col in range(5):
        indice = np.random.randint(X_test.shape[0])
        axs[row, col].imshow(X_test[indice], cmap="gray")
        axs[row, col].axis('off')
        axs[row, col].set_title('{} Pr√©dit {}'.format(dict[y_test[indice]], dict[y_pred[indice][0]]))
plt.show()
```
